```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (update filename if different)
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Attrition', data=df)
plt.title('Count of Employee Attrition')
plt.savefig('attrition_overall.png')
plt.show()
```

```python
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title('Attrition Count by Department')
plt.xticks(rotation=45)
plt.savefig('attrition_by_department.png')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='JobRole', hue='Attrition', data=df)
plt.title('Attrition Count by Job Role')
plt.xticks(rotation=45)
plt.savefig('attrition_by_jobrole.png')
plt.show()
```

* Most attrition is observed in the Sales department.
* Job roles like Laboratory Technician and Sales Executive show higher attrition.
* The company has significantly more employees staying than leaving.
